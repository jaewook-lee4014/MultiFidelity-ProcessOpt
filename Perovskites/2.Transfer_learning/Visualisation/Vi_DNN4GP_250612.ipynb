```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
# results_df.to_csv('transfer_learning_costs.csv', index=False)
# print("\nAll runs completed. Results saved to 'transfer_learning_costs.csv'")

results_df = pd.read_csv('../Pure_TL/transfer_learning_costs.csv')
# Create single figure with two plots side by side
plt.figure(figsize=(12, 6))

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Box plot
ax1.boxplot(results_df['total_cost'], patch_artist=True)
ax1.set_title('Box Plot')
ax1.set_ylabel('Cumulative cost [a.u.]')
ax1.set_ylim(0, 51)
# ax1.grid(True, axis='y', alpha=0.3)

# Violin plot
sns.violinplot(data=results_df['total_cost'], inner='box', color='lightblue', ax=ax2)
ax2.scatter(np.random.normal(0, 0.1, len(results_df['total_cost'])), 
           results_df['total_cost'], alpha=0.3, color='blue', label='Individual runs')
ax2.set_title('Violin Plot')
ax2.set_ylabel('Cumulative cost [a.u.]')
ax2.set_ylim(0, 51)
# ax2.grid(True, axis='y', alpha=0.3)
ax2.legend()

plt.suptitle('Distribution of Total Cost from 100 Runs')
plt.tight_layout()
plt.show()
```
